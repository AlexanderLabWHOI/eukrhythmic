```python
merged_assemblies = os.path.join("/", "vortexfs1", "omics", "alexander", "ncohen", "BATS2019-clio-metaT", "20Apr2020", "merged")

listedfiles = os.listdir(merged_assemblies)
sizes = [(os.path.join(merged_assemblies, path), os.stat(os.path.join(merged_assemblies, path)).st_size) for path in listedfiles]

## Grab the file sizes and translate them to an easy-to-understand form: 

filesizes = [float(curr[1]) / (1*10**9) for curr in sizes]
filenames = [curr[0].split("/")[len(curr[0].split("/")) - 1] for curr in sizes]
mergednottrans = pd.DataFrame({"Name": filenames, "Size in GB": filesizes})
mergednottrans
```
